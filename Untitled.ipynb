```python
# ======================================================================
# Cell 3: Flask æ‡‰ç”¨ç¨‹å¼ç¢¼ (åŠ å…¥æ¸…é™¤åŠŸèƒ½ + ä¿®æ­£é ç®—é‚è¼¯ + pyngrok)
# ======================================================================
import os
import pandas as pd
from flask import Flask, request, render_template_string, session # å°å…¥ session
from pyngrok import ngrok
import re
import atexit
import traceback
from thefuzz import process, fuzz # å°å…¥ thefuzz

# --- è¨­å®š ---
CSV_FILENAME = 'cars_data_expanded.csv'
# --- æ¬„ä½åç¨±å¸¸é‡ (ä¿æŒä¸è®Š) ---
BRAND_COL = 'å» ç‰Œ'; MODEL_COL = 'å‹è™Ÿ'; TRIM_COL = 'ç­‰ç´š'; YEAR_COL = 'å¹´ä»½'
BODY_COL = 'è»Šèº«å‹å¼'; ENGINE_COL = 'å‹•åŠ›'; FUEL_COL = 'æ²¹è€—åƒè€ƒ(km/L)'; SEATS_COL = 'åº§ä½æ•¸'
MIN_PRICE_COL = 'æœ€ä½åƒè€ƒåƒ¹(è¬)'; MAX_PRICE_COL = 'æœ€é«˜åƒè€ƒåƒ¹(è¬)'
TAGS_COL = 'ç‰¹è‰²æ¨™ç±¤'; LINK_COL = 'åƒè€ƒé€£çµ'
TAGS_LIST_COL = 'ç‰¹è‰²æ¨™ç±¤_list'
# --- å…¶ä»–è¨­å®š (ä¿æŒä¸è®Š) ---
DEFAULT_RESPONSE = "æŠ±æ­‰ï¼Œæˆ‘ä¸å¤ªæ˜ç™½ä½ çš„æ„æ€ã€‚ä½ å¯ä»¥è©¦è©¦å•ï¼š\n" \
                   "- 'Toyota'\n" \
                   "- 'é ç®— 100 è¬çš„ SUV'\n" \
                   "- 'çœæ²¹çš„æ²¹é›»è»Š'\n" \
                   "- 'æ¯”è¼ƒ Altis å’Œ Focus'\n" \
                   "- 'æœ‰å“ªäº›å» ç‰Œ'\n" \
                   "- 'help'\n" \
                   "- 'æ¸…é™¤ç´€éŒ„'" # åŠ å…¥æç¤º
MAX_RESULTS = 10
FUZZY_MATCH_THRESHOLD = 75
CLEAR_COMMANDS = ["æ¸…é™¤ç´€éŒ„", "æ¸…é™¤æ­·å²", "æ¸…é™¤å°è©±", "clear chat", "clear history", "reset chat", "æ¸…é™¤"] # æ¸…é™¤æŒ‡ä»¤é—œéµå­—

# --- å…¨åŸŸè®Šæ•¸ ---
app = Flask(__name__)
app.secret_key = 'a_very_secret_key_for_session' # Session éœ€è¦å¯†é‘°
car_data = pd.DataFrame()
known_brands_map = {}
known_body_types = []
known_engine_types = []
all_tags = set()

# --- åˆå§‹æ­¡è¿è¨Šæ¯ ---
INITIAL_BOT_MESSAGE = {
    "sender": "bot",
    "text": "ä½ å¥½ï¼æˆ‘æ˜¯é€²éšæ±½è»Šè³‡è¨ŠèŠå¤©æ©Ÿå™¨äººã€‚\n"
            "ä½ å¯ä»¥å•æˆ‘å„ç¨®æ¢ä»¶çµ„åˆ (ä¾‹å¦‚ 'é ç®—100è¬æœ‰ACCçš„æ²¹é›»SUV')ã€"
            "æ¯”è¼ƒè»Šæ¬¾ (ä¾‹å¦‚ 'æ¯”è¼ƒ CR-V å’Œ Kuga') æˆ–è¦æ±‚æ’åºã€‚\n"
            "è¼¸å…¥ 'help' æŸ¥çœ‹æ›´å¤šæŒ‡ä»¤ï¼Œè¼¸å…¥ 'æ¸…é™¤ç´€éŒ„' å¯æ¸…ç©ºå°è©±ã€‚"
}

# --- è³‡æ–™è¼‰å…¥ (`load_data` å‡½æ•¸ä¿æŒä¸è®Š) ---
def load_data():
    global car_data, known_brands_map, known_body_types, known_engine_types, all_tags
    try:
        colab_csv_path = os.path.join('/content/', CSV_FILENAME)
        print(f"Colab: æ­£åœ¨å˜—è©¦å¾ä»¥ä¸‹è·¯å¾‘è¼‰å…¥ CSV: {colab_csv_path}")
        if not os.path.exists(colab_csv_path):
            print(f"éŒ¯èª¤ï¼šåœ¨ Colab è·¯å¾‘æ‰¾ä¸åˆ° CSV æª”æ¡ˆ {colab_csv_path}")
            return False

        car_data = pd.read_csv(colab_csv_path)
        print(f"è®€å– {len(car_data)} ç­†åŸå§‹è³‡æ–™...")

        # --- é©—è­‰å’ŒåŸºæœ¬æ¸…ç† ---
        required_columns = [BRAND_COL, MODEL_COL, TRIM_COL, YEAR_COL, BODY_COL, ENGINE_COL,
                            FUEL_COL, SEATS_COL, MIN_PRICE_COL, MAX_PRICE_COL, TAGS_COL, LINK_COL]
        if not all(col in car_data.columns for col in required_columns):
            missing = [col for col in required_columns if col not in car_data.columns]
            print(f"éŒ¯èª¤ï¼šCSV ç¼ºå°‘å¿…è¦æ¬„ä½: {missing}")
            return False

        # --- è½‰æ›æ•¸å€¼é¡å‹ ---
        for col in [MIN_PRICE_COL, MAX_PRICE_COL, FUEL_COL, SEATS_COL, YEAR_COL]:
            car_data[col] = pd.to_numeric(car_data[col], errors='coerce')

        # --- è™•ç†ç‰¹è‰²æ¨™ç±¤ ---
        car_data[TAGS_COL] = car_data[TAGS_COL].fillna('')
        car_data[TAGS_LIST_COL] = car_data[TAGS_COL].apply(
            lambda x: [tag.strip().lower() for tag in x.split(',') if tag.strip()]
        )
        all_tags = set(tag for tags_list in car_data[TAGS_LIST_COL] for tag in tags_list)

        # --- ç§»é™¤åƒ¹æ ¼æˆ–å…¶ä»–é—œéµæ•¸å€¼ç‚º NaN çš„è¡Œ ---
        car_data.dropna(subset=[MIN_PRICE_COL, MAX_PRICE_COL, YEAR_COL], inplace=True)
        print(f"æ¸…ç†å¾Œå‰©é¤˜ {len(car_data)} ç­†æœ‰æ•ˆè³‡æ–™ã€‚")

        # --- ç²å–å·²çŸ¥åˆ†é¡è³‡è¨Š ---
        known_brands_map = {brand.lower(): brand for brand in car_data[BRAND_COL].unique()}
        known_body_types = [t.lower() for t in car_data[BODY_COL].unique()]
        known_engine_types = [t.lower() for t in car_data[ENGINE_COL].unique()]

        print(f"å·²çŸ¥å» ç‰Œ ({len(known_brands_map)}): {', '.join(known_brands_map.values())}")
        # print(f"å·²çŸ¥è»Šèº«å‹å¼: {', '.join(known_body_types)}") # å¯é¸æ‰“å°
        # print(f"å·²çŸ¥å‹•åŠ›é¡å‹: {', '.join(known_engine_types)}") # å¯é¸æ‰“å°
        return True

    except Exception as e:
        print(f"è®€å–æˆ–è™•ç† CSV æ™‚ç™¼ç”ŸéŒ¯èª¤ï¼š{e}")
        traceback.print_exc()
        return False


# --- è¼”åŠ©å‡½æ•¸ (`parse_budget`, `find_closest_match`, `format_car_details` ä¿æŒä¸è®Š) ---
def parse_budget(text):
    match = re.search(r'(\d+(\.\d+)?)\s*è¬', text)
    if match:
        try: return float(match.group(1))
        except ValueError: return None
    return None

def find_closest_match(text, choices, score_cutoff=FUZZY_MATCH_THRESHOLD):
    match = process.extractOne(text, choices, scorer=fuzz.WRatio, score_cutoff=score_cutoff)
    return match[0] if match else None

def format_car_details(row):
    details = f"- {row[YEAR_COL]:.0f} {row[BRAND_COL]} {row[MODEL_COL]} ({row[TRIM_COL]})\n"
    details += f"  é¡å‹: {row[BODY_COL]}, {row[ENGINE_COL]}, {row[SEATS_COL]:.0f}äººåº§\n"
    details += f"  åƒ¹æ ¼: ç´„ {row[MIN_PRICE_COL]:.0f}-{row[MAX_PRICE_COL]:.0f} è¬\n"
    if pd.notna(row[FUEL_COL]):
        unit = "km/L" if row[ENGINE_COL] != "ç´”é›»" else "sec 0-100km/h"
        details += f"  æ²¹è€—/èƒ½è€—åƒè€ƒ: {row[FUEL_COL]} {unit}\n"
    if row[TAGS_LIST_COL]:
        details += f"  ç‰¹è‰²: {', '.join(row[TAGS_LIST_COL])}\n"
    if pd.notna(row[LINK_COL]) and row[LINK_COL]:
        details += f"  é€£çµ: {row[LINK_COL]}\n"
    return details


# --- è§£æè¼¸å…¥ & ç”Ÿæˆå›æ‡‰ (åŠ å…¥æ¸…é™¤é‚è¼¯) ---
def get_response(user_input):
    global car_data, known_brands_map, known_body_types, known_engine_types, all_tags
    cleaned_input = user_input.strip().lower()
    original_input = user_input.strip()

    # --- *** æ–°å¢ï¼šæª¢æŸ¥æ¸…é™¤æŒ‡ä»¤ *** ---
    if cleaned_input in CLEAR_COMMANDS:
        if 'chat_history' in session:
            session['chat_history'] = [INITIAL_BOT_MESSAGE] # é‡ç½®ç‚ºåªæœ‰åˆå§‹è¨Šæ¯
            session.modified = True
            # æ¸…é™¤ä¸Šä¸‹æ–‡è¨˜æ†¶
            session.pop('last_brand', None)
            return "èŠå¤©ç´€éŒ„å·²æ¸…é™¤ã€‚"
        else:
            return "ç›®å‰æ²’æœ‰èŠå¤©ç´€éŒ„å¯ä»¥æ¸…é™¤ã€‚" # ç†è«–ä¸Šä¸æœƒç™¼ç”Ÿ

    # --- åŸºç¤æŒ‡ä»¤ (help, hello, list brands - ä¿æŒä¸è®Š) ---
    if cleaned_input in ["help", "å¹«åŠ©", "æŒ‡ä»¤"]:
        # (help å…§å®¹ä¿æŒä¸è®Š) ...
         return "ä½ å¯ä»¥è©¦è‘—è¼¸å…¥ï¼š\n" \
               "- å» ç‰Œåç¨± (ä¾‹å¦‚ 'Toyota', 'é ­åˆå¤§')\n" \
               "- é ç®— (ä¾‹å¦‚ 'é ç®— 80 è¬', '150è¬ä»¥ä¸‹')\n" \
               "- è»Šèº«å‹å¼ (ä¾‹å¦‚ 'SUV', 'è½è»Š')\n" \
               "- å‹•åŠ›é¡å‹ (ä¾‹å¦‚ 'æ²¹é›»', 'ç´”é›»')\n" \
               "- ç‰¹è‰² (ä¾‹å¦‚ 'ACC', 'Level 2', 'çœæ²¹')\n" \
               "- çµ„åˆæ¢ä»¶ (ä¾‹å¦‚ 'é ç®—100è¬æœ‰ACCçš„SUV')\n" \
               "- æ¯”è¼ƒ (ä¾‹å¦‚ 'æ¯”è¼ƒ Altis å’Œ Focus')\n" \
               "- æ’åº (ä¾‹å¦‚ 'Toyota æŒ‰åƒ¹æ ¼æ’åº', 'SUV å¾è²´åˆ°ä¾¿å®œ')\n" \
               "- 'æœ‰å“ªäº›å» ç‰Œ'\n" \
               "- 'æ¸…é™¤ç´€éŒ„'" # æ›´æ–° help å…§å®¹
    if cleaned_input in ["ä½ å¥½", "å—¨", "hello", "hi"]:
        return "ä½ å¥½ï¼è«‹å•éœ€è¦æŸ¥è©¢ä»€éº¼æ±½è»Šè³‡è¨Šå‘¢ï¼Ÿ (è¼¸å…¥ 'help' æŸ¥çœ‹æ›´å¤šæŒ‡ä»¤)"
    if cleaned_input in ["æœ‰å“ªäº›å» ç‰Œ", "å» ç‰Œåˆ—è¡¨", "æ‰€æœ‰å» ç‰Œ", "brands"]:
        if not car_data.empty:
            return "ç›®å‰è³‡æ–™åº«æ”¶éŒ„çš„å» ç‰Œæœ‰ï¼š\n- " + "\n- ".join(known_brands_map.values())
        else: return "æŠ±æ­‰ï¼Œç›®å‰ç„¡æ³•è¼‰å…¥å» ç‰Œè³‡æ–™ã€‚"

    # --- æ„åœ–åˆ†æèˆ‡å¯¦é«”æŠ½å– (ä¿æŒä¸è®Š) ---
    budget = parse_budget(cleaned_input)
    filters = {}
    sort_criteria = None
    compare_cars = None
    potential_brand = None

    compare_match = re.match(r'(æ¯”è¼ƒ|compare)\s*([\w\s-]+?)\s*(å’Œ|èˆ‡|è·Ÿ|and|vs)\s*([\w\s-]+)', cleaned_input)
    if compare_match:
        car1_name = compare_match.group(2).strip().lower()
        car2_name = compare_match.group(4).strip().lower()
        compare_cars = (car1_name, car2_name)
    else:
        # (æå– budget, brand, body, engine, tags, sort_criteria çš„é‚è¼¯ä¿æŒä¸è®Š) ...
        if budget: filters['budget'] = budget

        potential_brand_text = cleaned_input
        matched_brand_lower = find_closest_match(potential_brand_text, known_brands_map.keys())
        if matched_brand_lower:
            filters['brand'] = known_brands_map[matched_brand_lower]
            potential_brand = filters['brand']
        elif 'last_brand' in session and session['last_brand']:
            filters['brand'] = session['last_brand']

        for body_type in known_body_types:
            if body_type in cleaned_input: filters['body'] = body_type.capitalize()
        for engine_type in known_engine_types:
            if engine_type in cleaned_input: filters['engine'] = engine_type.capitalize()
        found_tags = [tag for tag in all_tags if tag in cleaned_input]
        if found_tags: filters['tags'] = found_tags

        if "æŒ‰åƒ¹æ ¼æ’åº" in cleaned_input or "å¾ä¾¿å®œåˆ°è²´" in cleaned_input:
            sort_criteria = {'by': MIN_PRICE_COL, 'ascending': True}
        elif "æŒ‰åƒ¹æ ¼å¾é«˜åˆ°ä½" in cleaned_input or "å¾è²´åˆ°ä¾¿å®œ" in cleaned_input:
            sort_criteria = {'by': MIN_PRICE_COL, 'ascending': False}
        elif "æŒ‰å¹´ä»½æ’åº" in cleaned_input or "å¾æ–°åˆ°èˆŠ" in cleaned_input:
            sort_criteria = {'by': YEAR_COL, 'ascending': False}
        elif "æŒ‰æ²¹è€—æ’åº" in cleaned_input or "æœ€çœæ²¹" in cleaned_input:
             sort_criteria = {'by': FUEL_COL, 'ascending': True}


    # --- æ ¹æ“šæ„åœ–åŸ·è¡Œå‹•ä½œ (æ¯”è¼ƒæˆ–ç¯©é¸é‚è¼¯ä¿æŒä¸è®Š) ---
    if compare_cars:
        # (æ¯”è¼ƒé‚è¼¯ä¿æŒä¸è®Š) ...
        response = f"æ­£åœ¨æ¯”è¼ƒ {compare_cars[0]} å’Œ {compare_cars[1]}:\n\n"
        found_count = 0
        comparison_details = ""
        for car_name in compare_cars:
            car_results = car_data[car_data[MODEL_COL].str.lower().str.contains(car_name)]
            if not car_results.empty:
                 car_row = car_results.iloc[0]
                 comparison_details += f"--- {car_row[BRAND_COL]} {car_row[MODEL_COL]} ({car_row[TRIM_COL]}) ---\n"
                 comparison_details += format_car_details(car_row) + "\n"
                 found_count += 1
            else:
                 comparison_details += f"--- æ‰¾ä¸åˆ°é—œæ–¼ {car_name} çš„è©³ç´°è³‡æ–™ ---\n\n"
        if found_count > 0: response += comparison_details
        else: response = f"æŠ±æ­‰ï¼Œæ‰¾ä¸åˆ°æ‚¨æƒ³æ¯”è¼ƒçš„è»Šæ¬¾ï¼š{compare_cars[0]} æˆ– {compare_cars[1]}ã€‚"

    elif filters:
        # (ç¯©é¸å’Œæ’åºé‚è¼¯ä¿æŒä¸è®Š) ...
        filtered_df = car_data.copy()
        filter_desc = []

        if 'budget' in filters:
            filtered_df = filtered_df[filtered_df[MIN_PRICE_COL] <= filters['budget']]
            filter_desc.append(f"é ç®— {filters['budget']:.0f} è¬å…§")
        if 'brand' in filters:
            filtered_df = filtered_df[filtered_df[BRAND_COL] == filters['brand']]
            filter_desc.append(f"å» ç‰Œ {filters['brand']}")
            session['last_brand'] = filters['brand']
        if 'body' in filters:
            filtered_df = filtered_df[filtered_df[BODY_COL].str.lower() == filters['body'].lower()]
            filter_desc.append(f"è»Šèº« {filters['body']}")
        if 'engine' in filters:
            filtered_df = filtered_df[filtered_df[ENGINE_COL].str.lower() == filters['engine'].lower()]
            filter_desc.append(f"å‹•åŠ› {filters['engine']}")
        if 'tags' in filters:
            for tag in filters['tags']:
                filtered_df = filtered_df[filtered_df[TAGS_LIST_COL].apply(lambda L: tag in L)]
            filter_desc.append(f"åŒ…å«ç‰¹è‰² '{'ã€'.join(filters['tags'])}'")

        if sort_criteria:
            filtered_df = filtered_df.sort_values(by=sort_criteria['by'], ascending=sort_criteria['ascending'], na_position='last')
            sort_desc = f"æŒ‰ {'åƒ¹æ ¼' if sort_criteria['by'] == MIN_PRICE_COL else ('å¹´ä»½' if sort_criteria['by'] == YEAR_COL else 'æ²¹è€—')} {'ç”±ä½åˆ°é«˜' if sort_criteria['ascending'] else 'ç”±é«˜åˆ°ä½'} æ’åº"
            filter_desc.append(sort_desc)

        if not filtered_df.empty:
            total_found = len(filtered_df)
            response = f"æ ¹æ“šæ¢ä»¶ [{', '.join(filter_desc)}]ï¼Œç‚ºæ‚¨æ‰¾åˆ° {total_found} ç­†çµæœ"
            if total_found > MAX_RESULTS:
                response += f" (åƒ…é¡¯ç¤ºå‰ {MAX_RESULTS} ç­†)ï¼š\n"
                filtered_df = filtered_df.head(MAX_RESULTS)
            else: response += "ï¼š\n"
            response += "--------------------\n"
            for index, row in filtered_df.iterrows(): response += format_car_details(row) + "\n"
            response += "--------------------"
        else:
            response = f"æŠ±æ­‰ï¼Œæ ¹æ“šæ¢ä»¶ [{', '.join(filter_desc)}]ï¼Œæ‰¾ä¸åˆ°ç¬¦åˆçš„è»Šæ¬¾ã€‚"
            if 'brand' not in filters and 'last_brand' in session and session['last_brand']:
                 response += f"\n(æ‚¨ä¸Šæ¬¡æŸ¥è©¢çš„å» ç‰Œæ˜¯ {session['last_brand']}ï¼Œå¦‚æœæƒ³æŸ¥åˆ¥çš„å» ç‰Œè«‹æ˜ç¢ºæŒ‡å®š)"
            session.pop('last_brand', None)

    else:
        response = DEFAULT_RESPONSE
        session.pop('last_brand', None)

    # æ›´æ–°ä¸Šä¸‹æ–‡è¨˜æ†¶ (ä¿æŒä¸è®Š)
    if potential_brand: session['last_brand'] = potential_brand
    elif 'brand' not in filters: session.pop('last_brand', None)

    return response


# --- Flask è·¯ç”±èˆ‡ HTML (home å‡½æ•¸ä¿®æ”¹ç‚ºä½¿ç”¨ INITIAL_BOT_MESSAGE) ---
HTML_TEMPLATE = """
<!doctype html>
<html>
<head>
    <title>é€²éšæ±½è»Šè³‡è¨ŠèŠå¤©æ©Ÿå™¨äºº</title>
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <style> /* CSS æ¨£å¼ä¿æŒä¸è®Š */
        body { font-family: sans-serif; margin: 20px; background-color: #f4f4f4; }
        .chatbox { max-width: 700px; margin: auto; background: #fff; padding: 20px; border-radius: 8px; box-shadow: 0 2px 5px rgba(0,0,0,0.1); }
        .history { height: 500px; overflow-y: scroll; border: 1px solid #ccc; padding: 10px; margin-bottom: 15px; background-color: #e9e9e9; border-radius: 4px; }
        .message { margin-bottom: 10px; padding: 8px 12px; border-radius: 5px; line-height: 1.4; }
        .user { background-color: #d1e7fd; text-align: right; margin-left: 50px; }
        .bot { background-color: #f8d7da; text-align: left; margin-right: 50px; white-space: pre-wrap; }
        .input-area { display: flex; margin-top: 10px; }
        input[type=text] { flex-grow: 1; padding: 10px; border: 1px solid #ccc; border-radius: 4px; }
        input[type=submit] { padding: 10px 15px; background-color: #0d6efd; color: white; border: none; border-radius: 4px; margin-left: 10px; cursor: pointer; }
        input[type=submit]:hover { background-color: #0b5ed7; }
        h2 { text-align: center; color: #333; }
        small { color: #666; display: block; text-align: center; margin-top: -10px; margin-bottom: 5px;}
        .hints { font-size: 0.85em; color: #555; text-align: center; margin-bottom: 15px; }
    </style>
</head>
<body>
    <div class="chatbox">
        <h2>é€²éšæ±½è»Šè³‡è¨ŠèŠå¤©æ©Ÿå™¨äºº</h2>
        <small>åƒ¹æ ¼åƒ…ç‚ºç²—ç•¥åƒè€ƒï¼Œéå¯¦éš›å ±åƒ¹</small>
        <div class="hints">
             è©¦è©¦å•: "é ç®—100è¬çš„SUV", "çœæ²¹", "æ¯”è¼ƒ Altis Focus", "Toyota æŒ‰åƒ¹æ ¼æ’åº", "æ¸…é™¤ç´€éŒ„", "help"
        </div>
        <div class="history">
            {% for entry in history %}
                {% if entry.sender == 'user' %}
                    <div class="message user">{{ entry.text }}</div>
                {% else %}
                    <div class="message bot">{{ entry.text }}</div>
                {% endif %}
            {% endfor %}
            <div id="anchor"></div>
        </div>
        <form method="post" class="input-area">
            <input type="text" name="user_input" placeholder="è¼¸å…¥æ¢ä»¶ã€æ¯”è¼ƒã€æ¸…é™¤ç´€éŒ„æˆ– 'help'..." autofocus>
            <input type="submit" value="ç™¼é€">
        </form>
    </div>
    <script>
        var chatHistory = document.querySelector('.history');
        chatHistory.scrollTop = chatHistory.scrollHeight;
    </script>
</body>
</html>
"""

@app.route('/', methods=['GET', 'POST'])
def home():
    if 'chat_history' not in session:
        session['chat_history'] = [INITIAL_BOT_MESSAGE] # ä½¿ç”¨å®šç¾©å¥½çš„åˆå§‹è¨Šæ¯
        session.modified = True

    if request.method == 'POST':
        user_input = request.form['user_input']
        if user_input:
            # å…ˆå°‡ä½¿ç”¨è€…è¼¸å…¥åŠ åˆ°æ­·å²
            session['chat_history'].append({"sender": "user", "text": user_input})
            session.modified = True # æ¨™è¨˜ä¿®æ”¹

            # ç²å–æ©Ÿå™¨äººå›æ‡‰ (get_response å…§éƒ¨å¯èƒ½æœƒä¿®æ”¹ session['chat_history'])
            bot_response = get_response(user_input)

            # åªæœ‰åœ¨ä¸æ˜¯æ¸…é™¤æŒ‡ä»¤æˆåŠŸæ™‚æ‰å°‡æ©Ÿå™¨äººå›æ‡‰åŠ åˆ°æ­·å²
            # (å› ç‚ºæ¸…é™¤æŒ‡ä»¤çš„å›æ‡‰åªæ˜¯ç¢ºèªè¨Šæ¯ï¼Œå¯¦éš›æ­·å²å·²è¢« get_response é‡ç½®)
            if user_input.strip().lower() not in CLEAR_COMMANDS:
                 session['chat_history'].append({"sender": "bot", "text": bot_response})

            # é™åˆ¶æ­·å²ç´€éŒ„é•·åº¦
            if len(session['chat_history']) > 30:
                session['chat_history'] = session['chat_history'][-30:]
            session.modified = True # å†æ¬¡æ¨™è¨˜ä¿®æ”¹

    return render_template_string(HTML_TEMPLATE, history=session.get('chat_history', []))


# --- å•Ÿå‹•æ‡‰ç”¨ (pyngrok é‚è¼¯ä¿æŒä¸è®Š) ---
if __name__ == '__main__':
    if load_data():
        print("è³‡æ–™è¼‰å…¥æˆåŠŸ...")
        try:
            ngrok.kill()
            # --- é‡è¦ï¼šè¨­ç½®ä½ çš„ Ngrok Authtoken ---
            NGROK_AUTH_TOKEN = "YOUR_NGROK_AUTH_TOKEN" # <--- åœ¨é€™è£¡è²¼ä¸Šä½ çš„ Token
            if NGROK_AUTH_TOKEN != "YOUR_NGROK_AUTH_TOKEN":
               ngrok.set_auth_token(NGROK_AUTH_TOKEN)
            else:
               print("æé†’ï¼šå»ºè­°è¨­å®š ngrok Authtokenã€‚")

            public_url = ngrok.connect(5000)
            print("*" * 50)
            print(f"âœ… Ngrok é€šé“å·²å»ºç«‹ï¼")
            print(f"ğŸ”— å…¬é–‹è¨ªå•ç¶²å€: {public_url}")
            print("*" * 50)
            app.config['BASE_URL'] = public_url
            atexit.register(ngrok.kill)

            print("æ­£åœ¨å•Ÿå‹• Flask æ‡‰ç”¨...")
            app.run(port=5000)

        except Exception as e:
            print(f"âŒ å•Ÿå‹• ngrok æˆ– Flask æ™‚ç™¼ç”ŸéŒ¯èª¤:")
            traceback.print_exc()
            print("\nğŸ’¡ æç¤ºï¼šè«‹æª¢æŸ¥ ngrok Authtoken æˆ–å˜—è©¦é‡æ–°å•Ÿå‹• Runtimeã€‚")
    else:
        print("âŒ è³‡æ–™è¼‰å…¥å¤±æ•—ï¼Œç„¡æ³•å•Ÿå‹•èŠå¤©æ©Ÿå™¨äººã€‚è«‹æª¢æŸ¥ CSV æª”æ¡ˆã€‚")
```
