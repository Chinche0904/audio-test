```python
# ======================================================================
# Cell 3: Flask 應用程式碼 (加入清除功能 + 修正預算邏輯 + pyngrok)
# ======================================================================
import os
import pandas as pd
from flask import Flask, request, render_template_string, session # 導入 session
from pyngrok import ngrok
import re
import atexit
import traceback
from thefuzz import process, fuzz # 導入 thefuzz

# --- 設定 ---
CSV_FILENAME = 'cars_data_expanded.csv'
# --- 欄位名稱常量 (保持不變) ---
BRAND_COL = '廠牌'; MODEL_COL = '型號'; TRIM_COL = '等級'; YEAR_COL = '年份'
BODY_COL = '車身型式'; ENGINE_COL = '動力'; FUEL_COL = '油耗參考(km/L)'; SEATS_COL = '座位數'
MIN_PRICE_COL = '最低參考價(萬)'; MAX_PRICE_COL = '最高參考價(萬)'
TAGS_COL = '特色標籤'; LINK_COL = '參考連結'
TAGS_LIST_COL = '特色標籤_list'
# --- 其他設定 (保持不變) ---
DEFAULT_RESPONSE = "抱歉，我不太明白你的意思。你可以試試問：\n" \
                   "- 'Toyota'\n" \
                   "- '預算 100 萬的 SUV'\n" \
                   "- '省油的油電車'\n" \
                   "- '比較 Altis 和 Focus'\n" \
                   "- '有哪些廠牌'\n" \
                   "- 'help'\n" \
                   "- '清除紀錄'" # 加入提示
MAX_RESULTS = 10
FUZZY_MATCH_THRESHOLD = 75
CLEAR_COMMANDS = ["清除紀錄", "清除歷史", "清除對話", "clear chat", "clear history", "reset chat", "清除"] # 清除指令關鍵字

# --- 全域變數 ---
app = Flask(__name__)
app.secret_key = 'a_very_secret_key_for_session' # Session 需要密鑰
car_data = pd.DataFrame()
known_brands_map = {}
known_body_types = []
known_engine_types = []
all_tags = set()

# --- 初始歡迎訊息 ---
INITIAL_BOT_MESSAGE = {
    "sender": "bot",
    "text": "你好！我是進階汽車資訊聊天機器人。\n"
            "你可以問我各種條件組合 (例如 '預算100萬有ACC的油電SUV')、"
            "比較車款 (例如 '比較 CR-V 和 Kuga') 或要求排序。\n"
            "輸入 'help' 查看更多指令，輸入 '清除紀錄' 可清空對話。"
}

# --- 資料載入 (`load_data` 函數保持不變) ---
def load_data():
    global car_data, known_brands_map, known_body_types, known_engine_types, all_tags
    try:
        colab_csv_path = os.path.join('/content/', CSV_FILENAME)
        print(f"Colab: 正在嘗試從以下路徑載入 CSV: {colab_csv_path}")
        if not os.path.exists(colab_csv_path):
            print(f"錯誤：在 Colab 路徑找不到 CSV 檔案 {colab_csv_path}")
            return False

        car_data = pd.read_csv(colab_csv_path)
        print(f"讀取 {len(car_data)} 筆原始資料...")

        # --- 驗證和基本清理 ---
        required_columns = [BRAND_COL, MODEL_COL, TRIM_COL, YEAR_COL, BODY_COL, ENGINE_COL,
                            FUEL_COL, SEATS_COL, MIN_PRICE_COL, MAX_PRICE_COL, TAGS_COL, LINK_COL]
        if not all(col in car_data.columns for col in required_columns):
            missing = [col for col in required_columns if col not in car_data.columns]
            print(f"錯誤：CSV 缺少必要欄位: {missing}")
            return False

        # --- 轉換數值類型 ---
        for col in [MIN_PRICE_COL, MAX_PRICE_COL, FUEL_COL, SEATS_COL, YEAR_COL]:
            car_data[col] = pd.to_numeric(car_data[col], errors='coerce')

        # --- 處理特色標籤 ---
        car_data[TAGS_COL] = car_data[TAGS_COL].fillna('')
        car_data[TAGS_LIST_COL] = car_data[TAGS_COL].apply(
            lambda x: [tag.strip().lower() for tag in x.split(',') if tag.strip()]
        )
        all_tags = set(tag for tags_list in car_data[TAGS_LIST_COL] for tag in tags_list)

        # --- 移除價格或其他關鍵數值為 NaN 的行 ---
        car_data.dropna(subset=[MIN_PRICE_COL, MAX_PRICE_COL, YEAR_COL], inplace=True)
        print(f"清理後剩餘 {len(car_data)} 筆有效資料。")

        # --- 獲取已知分類資訊 ---
        known_brands_map = {brand.lower(): brand for brand in car_data[BRAND_COL].unique()}
        known_body_types = [t.lower() for t in car_data[BODY_COL].unique()]
        known_engine_types = [t.lower() for t in car_data[ENGINE_COL].unique()]

        print(f"已知廠牌 ({len(known_brands_map)}): {', '.join(known_brands_map.values())}")
        # print(f"已知車身型式: {', '.join(known_body_types)}") # 可選打印
        # print(f"已知動力類型: {', '.join(known_engine_types)}") # 可選打印
        return True

    except Exception as e:
        print(f"讀取或處理 CSV 時發生錯誤：{e}")
        traceback.print_exc()
        return False


# --- 輔助函數 (`parse_budget`, `find_closest_match`, `format_car_details` 保持不變) ---
def parse_budget(text):
    match = re.search(r'(\d+(\.\d+)?)\s*萬', text)
    if match:
        try: return float(match.group(1))
        except ValueError: return None
    return None

def find_closest_match(text, choices, score_cutoff=FUZZY_MATCH_THRESHOLD):
    match = process.extractOne(text, choices, scorer=fuzz.WRatio, score_cutoff=score_cutoff)
    return match[0] if match else None

def format_car_details(row):
    details = f"- {row[YEAR_COL]:.0f} {row[BRAND_COL]} {row[MODEL_COL]} ({row[TRIM_COL]})\n"
    details += f"  類型: {row[BODY_COL]}, {row[ENGINE_COL]}, {row[SEATS_COL]:.0f}人座\n"
    details += f"  價格: 約 {row[MIN_PRICE_COL]:.0f}-{row[MAX_PRICE_COL]:.0f} 萬\n"
    if pd.notna(row[FUEL_COL]):
        unit = "km/L" if row[ENGINE_COL] != "純電" else "sec 0-100km/h"
        details += f"  油耗/能耗參考: {row[FUEL_COL]} {unit}\n"
    if row[TAGS_LIST_COL]:
        details += f"  特色: {', '.join(row[TAGS_LIST_COL])}\n"
    if pd.notna(row[LINK_COL]) and row[LINK_COL]:
        details += f"  連結: {row[LINK_COL]}\n"
    return details


# --- 解析輸入 & 生成回應 (加入清除邏輯) ---
def get_response(user_input):
    global car_data, known_brands_map, known_body_types, known_engine_types, all_tags
    cleaned_input = user_input.strip().lower()
    original_input = user_input.strip()

    # --- *** 新增：檢查清除指令 *** ---
    if cleaned_input in CLEAR_COMMANDS:
        if 'chat_history' in session:
            session['chat_history'] = [INITIAL_BOT_MESSAGE] # 重置為只有初始訊息
            session.modified = True
            # 清除上下文記憶
            session.pop('last_brand', None)
            return "聊天紀錄已清除。"
        else:
            return "目前沒有聊天紀錄可以清除。" # 理論上不會發生

    # --- 基礎指令 (help, hello, list brands - 保持不變) ---
    if cleaned_input in ["help", "幫助", "指令"]:
        # (help 內容保持不變) ...
         return "你可以試著輸入：\n" \
               "- 廠牌名稱 (例如 'Toyota', '頭又大')\n" \
               "- 預算 (例如 '預算 80 萬', '150萬以下')\n" \
               "- 車身型式 (例如 'SUV', '轎車')\n" \
               "- 動力類型 (例如 '油電', '純電')\n" \
               "- 特色 (例如 'ACC', 'Level 2', '省油')\n" \
               "- 組合條件 (例如 '預算100萬有ACC的SUV')\n" \
               "- 比較 (例如 '比較 Altis 和 Focus')\n" \
               "- 排序 (例如 'Toyota 按價格排序', 'SUV 從貴到便宜')\n" \
               "- '有哪些廠牌'\n" \
               "- '清除紀錄'" # 更新 help 內容
    if cleaned_input in ["你好", "嗨", "hello", "hi"]:
        return "你好！請問需要查詢什麼汽車資訊呢？ (輸入 'help' 查看更多指令)"
    if cleaned_input in ["有哪些廠牌", "廠牌列表", "所有廠牌", "brands"]:
        if not car_data.empty:
            return "目前資料庫收錄的廠牌有：\n- " + "\n- ".join(known_brands_map.values())
        else: return "抱歉，目前無法載入廠牌資料。"

    # --- 意圖分析與實體抽取 (保持不變) ---
    budget = parse_budget(cleaned_input)
    filters = {}
    sort_criteria = None
    compare_cars = None
    potential_brand = None

    compare_match = re.match(r'(比較|compare)\s*([\w\s-]+?)\s*(和|與|跟|and|vs)\s*([\w\s-]+)', cleaned_input)
    if compare_match:
        car1_name = compare_match.group(2).strip().lower()
        car2_name = compare_match.group(4).strip().lower()
        compare_cars = (car1_name, car2_name)
    else:
        # (提取 budget, brand, body, engine, tags, sort_criteria 的邏輯保持不變) ...
        if budget: filters['budget'] = budget

        potential_brand_text = cleaned_input
        matched_brand_lower = find_closest_match(potential_brand_text, known_brands_map.keys())
        if matched_brand_lower:
            filters['brand'] = known_brands_map[matched_brand_lower]
            potential_brand = filters['brand']
        elif 'last_brand' in session and session['last_brand']:
            filters['brand'] = session['last_brand']

        for body_type in known_body_types:
            if body_type in cleaned_input: filters['body'] = body_type.capitalize()
        for engine_type in known_engine_types:
            if engine_type in cleaned_input: filters['engine'] = engine_type.capitalize()
        found_tags = [tag for tag in all_tags if tag in cleaned_input]
        if found_tags: filters['tags'] = found_tags

        if "按價格排序" in cleaned_input or "從便宜到貴" in cleaned_input:
            sort_criteria = {'by': MIN_PRICE_COL, 'ascending': True}
        elif "按價格從高到低" in cleaned_input or "從貴到便宜" in cleaned_input:
            sort_criteria = {'by': MIN_PRICE_COL, 'ascending': False}
        elif "按年份排序" in cleaned_input or "從新到舊" in cleaned_input:
            sort_criteria = {'by': YEAR_COL, 'ascending': False}
        elif "按油耗排序" in cleaned_input or "最省油" in cleaned_input:
             sort_criteria = {'by': FUEL_COL, 'ascending': True}


    # --- 根據意圖執行動作 (比較或篩選邏輯保持不變) ---
    if compare_cars:
        # (比較邏輯保持不變) ...
        response = f"正在比較 {compare_cars[0]} 和 {compare_cars[1]}:\n\n"
        found_count = 0
        comparison_details = ""
        for car_name in compare_cars:
            car_results = car_data[car_data[MODEL_COL].str.lower().str.contains(car_name)]
            if not car_results.empty:
                 car_row = car_results.iloc[0]
                 comparison_details += f"--- {car_row[BRAND_COL]} {car_row[MODEL_COL]} ({car_row[TRIM_COL]}) ---\n"
                 comparison_details += format_car_details(car_row) + "\n"
                 found_count += 1
            else:
                 comparison_details += f"--- 找不到關於 {car_name} 的詳細資料 ---\n\n"
        if found_count > 0: response += comparison_details
        else: response = f"抱歉，找不到您想比較的車款：{compare_cars[0]} 或 {compare_cars[1]}。"

    elif filters:
        # (篩選和排序邏輯保持不變) ...
        filtered_df = car_data.copy()
        filter_desc = []

        if 'budget' in filters:
            filtered_df = filtered_df[filtered_df[MIN_PRICE_COL] <= filters['budget']]
            filter_desc.append(f"預算 {filters['budget']:.0f} 萬內")
        if 'brand' in filters:
            filtered_df = filtered_df[filtered_df[BRAND_COL] == filters['brand']]
            filter_desc.append(f"廠牌 {filters['brand']}")
            session['last_brand'] = filters['brand']
        if 'body' in filters:
            filtered_df = filtered_df[filtered_df[BODY_COL].str.lower() == filters['body'].lower()]
            filter_desc.append(f"車身 {filters['body']}")
        if 'engine' in filters:
            filtered_df = filtered_df[filtered_df[ENGINE_COL].str.lower() == filters['engine'].lower()]
            filter_desc.append(f"動力 {filters['engine']}")
        if 'tags' in filters:
            for tag in filters['tags']:
                filtered_df = filtered_df[filtered_df[TAGS_LIST_COL].apply(lambda L: tag in L)]
            filter_desc.append(f"包含特色 '{'、'.join(filters['tags'])}'")

        if sort_criteria:
            filtered_df = filtered_df.sort_values(by=sort_criteria['by'], ascending=sort_criteria['ascending'], na_position='last')
            sort_desc = f"按 {'價格' if sort_criteria['by'] == MIN_PRICE_COL else ('年份' if sort_criteria['by'] == YEAR_COL else '油耗')} {'由低到高' if sort_criteria['ascending'] else '由高到低'} 排序"
            filter_desc.append(sort_desc)

        if not filtered_df.empty:
            total_found = len(filtered_df)
            response = f"根據條件 [{', '.join(filter_desc)}]，為您找到 {total_found} 筆結果"
            if total_found > MAX_RESULTS:
                response += f" (僅顯示前 {MAX_RESULTS} 筆)：\n"
                filtered_df = filtered_df.head(MAX_RESULTS)
            else: response += "：\n"
            response += "--------------------\n"
            for index, row in filtered_df.iterrows(): response += format_car_details(row) + "\n"
            response += "--------------------"
        else:
            response = f"抱歉，根據條件 [{', '.join(filter_desc)}]，找不到符合的車款。"
            if 'brand' not in filters and 'last_brand' in session and session['last_brand']:
                 response += f"\n(您上次查詢的廠牌是 {session['last_brand']}，如果想查別的廠牌請明確指定)"
            session.pop('last_brand', None)

    else:
        response = DEFAULT_RESPONSE
        session.pop('last_brand', None)

    # 更新上下文記憶 (保持不變)
    if potential_brand: session['last_brand'] = potential_brand
    elif 'brand' not in filters: session.pop('last_brand', None)

    return response


# --- Flask 路由與 HTML (home 函數修改為使用 INITIAL_BOT_MESSAGE) ---
HTML_TEMPLATE = """
<!doctype html>
<html>
<head>
    <title>進階汽車資訊聊天機器人</title>
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <style> /* CSS 樣式保持不變 */
        body { font-family: sans-serif; margin: 20px; background-color: #f4f4f4; }
        .chatbox { max-width: 700px; margin: auto; background: #fff; padding: 20px; border-radius: 8px; box-shadow: 0 2px 5px rgba(0,0,0,0.1); }
        .history { height: 500px; overflow-y: scroll; border: 1px solid #ccc; padding: 10px; margin-bottom: 15px; background-color: #e9e9e9; border-radius: 4px; }
        .message { margin-bottom: 10px; padding: 8px 12px; border-radius: 5px; line-height: 1.4; }
        .user { background-color: #d1e7fd; text-align: right; margin-left: 50px; }
        .bot { background-color: #f8d7da; text-align: left; margin-right: 50px; white-space: pre-wrap; }
        .input-area { display: flex; margin-top: 10px; }
        input[type=text] { flex-grow: 1; padding: 10px; border: 1px solid #ccc; border-radius: 4px; }
        input[type=submit] { padding: 10px 15px; background-color: #0d6efd; color: white; border: none; border-radius: 4px; margin-left: 10px; cursor: pointer; }
        input[type=submit]:hover { background-color: #0b5ed7; }
        h2 { text-align: center; color: #333; }
        small { color: #666; display: block; text-align: center; margin-top: -10px; margin-bottom: 5px;}
        .hints { font-size: 0.85em; color: #555; text-align: center; margin-bottom: 15px; }
    </style>
</head>
<body>
    <div class="chatbox">
        <h2>進階汽車資訊聊天機器人</h2>
        <small>價格僅為粗略參考，非實際報價</small>
        <div class="hints">
             試試問: "預算100萬的SUV", "省油", "比較 Altis Focus", "Toyota 按價格排序", "清除紀錄", "help"
        </div>
        <div class="history">
            {% for entry in history %}
                {% if entry.sender == 'user' %}
                    <div class="message user">{{ entry.text }}</div>
                {% else %}
                    <div class="message bot">{{ entry.text }}</div>
                {% endif %}
            {% endfor %}
            <div id="anchor"></div>
        </div>
        <form method="post" class="input-area">
            <input type="text" name="user_input" placeholder="輸入條件、比較、清除紀錄或 'help'..." autofocus>
            <input type="submit" value="發送">
        </form>
    </div>
    <script>
        var chatHistory = document.querySelector('.history');
        chatHistory.scrollTop = chatHistory.scrollHeight;
    </script>
</body>
</html>
"""

@app.route('/', methods=['GET', 'POST'])
def home():
    if 'chat_history' not in session:
        session['chat_history'] = [INITIAL_BOT_MESSAGE] # 使用定義好的初始訊息
        session.modified = True

    if request.method == 'POST':
        user_input = request.form['user_input']
        if user_input:
            # 先將使用者輸入加到歷史
            session['chat_history'].append({"sender": "user", "text": user_input})
            session.modified = True # 標記修改

            # 獲取機器人回應 (get_response 內部可能會修改 session['chat_history'])
            bot_response = get_response(user_input)

            # 只有在不是清除指令成功時才將機器人回應加到歷史
            # (因為清除指令的回應只是確認訊息，實際歷史已被 get_response 重置)
            if user_input.strip().lower() not in CLEAR_COMMANDS:
                 session['chat_history'].append({"sender": "bot", "text": bot_response})

            # 限制歷史紀錄長度
            if len(session['chat_history']) > 30:
                session['chat_history'] = session['chat_history'][-30:]
            session.modified = True # 再次標記修改

    return render_template_string(HTML_TEMPLATE, history=session.get('chat_history', []))


# --- 啟動應用 (pyngrok 邏輯保持不變) ---
if __name__ == '__main__':
    if load_data():
        print("資料載入成功...")
        try:
            ngrok.kill()
            # --- 重要：設置你的 Ngrok Authtoken ---
            NGROK_AUTH_TOKEN = "YOUR_NGROK_AUTH_TOKEN" # <--- 在這裡貼上你的 Token
            if NGROK_AUTH_TOKEN != "YOUR_NGROK_AUTH_TOKEN":
               ngrok.set_auth_token(NGROK_AUTH_TOKEN)
            else:
               print("提醒：建議設定 ngrok Authtoken。")

            public_url = ngrok.connect(5000)
            print("*" * 50)
            print(f"✅ Ngrok 通道已建立！")
            print(f"🔗 公開訪問網址: {public_url}")
            print("*" * 50)
            app.config['BASE_URL'] = public_url
            atexit.register(ngrok.kill)

            print("正在啟動 Flask 應用...")
            app.run(port=5000)

        except Exception as e:
            print(f"❌ 啟動 ngrok 或 Flask 時發生錯誤:")
            traceback.print_exc()
            print("\n💡 提示：請檢查 ngrok Authtoken 或嘗試重新啟動 Runtime。")
    else:
        print("❌ 資料載入失敗，無法啟動聊天機器人。請檢查 CSV 檔案。")
```
